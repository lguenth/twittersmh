# Tweets parsen und bereinigen

## Einlesen des Korpus

```python
import pandas as pd

corpus = "/home/lukel/projekte/twittersmh/data/tweepy_v2_corpus.csv"
df = pd.read_csv(corpus, sep=";", header=None, dtype=object, names = [
        "created_at",
        "tweet_id",
        "text",
        "conversation_id",
        "author_id",
        "entities",
        "public_metrics",
        "referenced_tweets",
        "lang",
    ])
```

## Datum formatieren und ungültige Einträge verwerfen

```python
df["created_at"] = pd.to_datetime(df["created_at"], errors="coerce")

print("These entries could not be converted:")
nan = df[df["created_at"].isna() == True]
nan

df["created_at"].dropna(inplace=True)
```

```python
df["day"] = df["created_at"].dt.date
df["time"] = df["created_at"].dt.time
df.sort_values(by="day", inplace=True, ignore_index=True)
df.head(5)
```

## Deutsche Tweets herausfiltern

```python
de = df[df["lang"] == "de"]
de = de.drop("lang", axis=1)
de.head(5)
```

## Retweets herausfiltern und in separater Tabelle speichern

```python
retweets = de[de["text"].str.contains("RT @")]
de = de[de["text"].str.contains("RT @") == False]
```

## Tweets bereinigen

```python
import re
import numpy as np

tweets_clean, emojis, hashtags, mentions, urls = [[] for _ in range(5)]

emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
            u"\U00002500-\U00002BEF"  # chinese char
            u"\U00002702-\U000027B0"
            u"\U00002702-\U000027B0"
            u"\U000024C2-\U0001F251"
            u"\U0001f926-\U0001f937"
            u"\U00010000-\U0010ffff"
            u"\u2640-\u2642"
            u"\u2600-\u2B55"
            u"\u200d"
            u"\u23cf"
            u"\u23e9"
            u"\u231a"
            u"\ufe0f"  # dingbats
            u"\u3030"
            "]+", flags=re.UNICODE)

url_pattern = re.compile("http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+") # This will match all kinds of URLs

for tweet in de["text"]:
    emoji = re.findall(emoji_pattern, tweet) # Extract emojis
    if not emoji:
        emoji = np.NaN # Replace with null value if empty
    tweet = re.sub(emoji_pattern, "", tweet) # Remove emojis from tweets
    
    hashtag = re.findall(r"#[A-Za-z0-9]+", tweet) # Extract hashtags
    if not hashtag:
        hashtag = np.NaN
    tweet = tweet.replace("#", "").replace("_", "") # Remove hashtag sign but keep the text

    mention = re.findall(r"@[A-Za-z0-9]+", tweet)
    if not mention:
        mention = np.NaN
    tweet = re.sub(r"@[A-Za-z0-9]+", "", tweet)
    
    url = re.findall(url_pattern, tweet) 
    if not url:
        url = np.NaN
    tweet = re.sub(url_pattern, "", tweet)

    tweet = " ".join(tweet.split())

    tweets_clean.append(tweet)
    emojis.append(emoji)
    hashtags.append(hashtag)
    mentions.append(mention)
    urls.append(url)

de["text_clean"], de["emojis"], de["hashtags"], de["mentions"], de["urls"] = tweets_clean, emojis, hashtags, mentions, urls
```
